```python
from pathlib import Path
from IPython.display import display, Image, HTML, Markdown
```

```python

# Get all images in current directory
notebook_dir = Path(".")
images = sorted(notebook_dir.glob("*.png"))

# Display title
display(Markdown("# Figure Gallery"))

# Start grid container
display(HTML("""
<style>
.gallery {
    display: grid;
    grid-template-columns: repeat(auto-fill, minmax(300px, 1fr));
    gap: 20px;
    padding: 20px 0;
}
.gallery-item {
    border: 1px solid #e0e0e0;
    border-radius: 8px;
    overflow: hidden;
    box-shadow: 0 2px 8px rgba(0,0,0,0.1);
    padding: 10px;
    text-align: center;
}
.gallery-caption {
    padding: 10px;
    font-size: 14px;
    font-weight: bold;
}
</style>
"""))

# Display each image - this embeds the actual image data
for img in images:
    caption = img.stem.replace("_", " ").title()
    
    display(HTML('<div class="gallery-item">'))
    # This embeds the actual image data as base64 in the notebook
    display(Image(filename=str(img)))
    display(HTML(f'<div class="gallery-caption">{caption}</div>'))
    display(HTML('</div>'))
```
